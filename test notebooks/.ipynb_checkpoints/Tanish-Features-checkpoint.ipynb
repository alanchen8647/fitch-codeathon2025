## Plan Overview
- **Scale & Governance**: capture non-linear revenue and ESG score gaps (env vs. overall, governance vs. env).
- **Sector Structure**: pivot NACE revenues, compute concentration metrics, and flag carbon-intensive mixes.
- **Environmental Actions**: summarize activity counts/adjustments to differentiate proactive emitters.
- **SDG Commitments**: encode SDG pledges (especially 2, 7, 12, 13) and bundle indicators.
- **Interactions**: combine scale with intensity (e.g., log-revenue × high-intensity share).
Visual validations below demonstrate why each block adds signal.

```python
from pathlib import Path
import numpy as np
import pandas as pd
from textwrap import fill
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-muted')
```

```python
DATA_DIR = Path('../data')
train = pd.read_csv('data/train.csv')
sect = pd.read_csv('data/revenue_distribution_by_sector.csv')
env = pd.read_csv('data/environmental_activities.csv')
sdg = pd.read_csv('data/sustainable_development_goals.csv')

def add_sector_features(df, sector):
    level1 = (
        sector.pivot_table(
            values='revenue_pct',
            index='entity_id',
            columns='nace_level_1_code',
            aggfunc='sum',
            fill_value=0
        )
        .add_prefix('sect_')
        .add_suffix('_pct')
    )
    pct = level1.to_numpy(dtype=float)
    totals = pct.sum(axis=1, keepdims=True)
    pct = np.divide(pct, totals, out=np.zeros_like(pct), where=totals != 0)
    with np.errstate(divide='ignore', invalid='ignore'):
        log_pct = np.where(pct > 0, np.log(pct), 0)
    level1['sect_hhi'] = (pct ** 2).sum(axis=1)
    level1['sect_entropy'] = -(pct * log_pct).sum(axis=1)
    level1['sect_top1_pct'] = np.sort(pct, axis=1)[:, -1]
    level1['sect_top3_pct'] = np.sort(pct, axis=1)[:, -3:].sum(axis=1)
    level1['sect_num_lvl1'] = (level1 > 0).sum(axis=1)
    level1['lvl2_count'] = sector.groupby('entity_id')['nace_level_2_code'].nunique()
    df = df.merge(level1.reset_index(), on='entity_id', how='left')
    high_intensity = ['sect_B_pct','sect_C_pct','sect_D_pct','sect_E_pct','sect_F_pct','sect_H_pct']
    for col in high_intensity:
        if col not in df:
            df[col] = 0.0
    df['sect_high_intensity_pct'] = df[high_intensity].sum(axis=1)
    return df.fillna(0)

def add_environmental_features(df, env):
    if env.empty:
        env_agg = pd.DataFrame({'entity_id': df['entity_id'].unique()})
        env_agg[['env_adj_sum','env_adj_avg','env_adj_min','env_adj_max','env_event_cnt','env_positive_cnt','env_negative_cnt']] = 0
    else:
        env_agg = (
            env.groupby('entity_id')
            .agg(
                env_adj_sum=('env_score_adjustment','sum'),
                env_adj_avg=('env_score_adjustment','mean'),
                env_adj_min=('env_score_adjustment','min'),
                env_adj_max=('env_score_adjustment','max'),
                env_event_cnt=('env_score_adjustment','size'),
                env_positive_cnt=('env_score_adjustment', lambda s: (s < 0).sum()),
                env_negative_cnt=('env_score_adjustment', lambda s: (s > 0).sum()),
            )
            .reset_index()
        )
        type_pivot = pd.get_dummies(env[['entity_id','activity_type']], columns=['activity_type']).groupby('entity_id').sum().reset_index()
        env_agg = env_agg.merge(type_pivot, on='entity_id', how='outer')
    return df.merge(env_agg.fillna(0), on='entity_id', how='left').fillna(0)

def add_sdg_features(df, sdg):
    if sdg.empty:
        base = pd.DataFrame({'entity_id': df['entity_id'].unique(), 'sdg_count':0, 'sdg_unique':0})
        for goal in [2,7,12,13]:
            base[f'sdg_{goal}_flag'] = 0
        base['sdg_has_climate_bundle'] = 0
        return df.merge(base, on='entity_id', how='left')
    sdg_flags = pd.get_dummies(sdg[['entity_id','sdg_id']], columns=['sdg_id'], prefix='sdg').groupby('entity_id').sum().clip(upper=1).reset_index()
    rename_map = {col: f"{col}_flag" for col in sdg_flags.columns if col != 'entity_id'}
    sdg_flags = sdg_flags.rename(columns=rename_map)
    sdg_counts = sdg.groupby('entity_id').agg(sdg_count=('sdg_id','size'), sdg_unique=('sdg_id','nunique')).reset_index()
    sdg_agg = sdg_counts.merge(sdg_flags, on='entity_id', how='outer').fillna(0)
    for goal in [2,7,12,13]:
        col = f'sdg_{goal}_flag'
        if col not in sdg_agg:
            sdg_agg[col] = 0
    seven = sdg_agg.get('sdg_7_flag', pd.Series(0, index=sdg_agg.index))
    thirteen = sdg_agg.get('sdg_13_flag', pd.Series(0, index=sdg_agg.index))
    sdg_agg['sdg_has_climate_bundle'] = ((seven > 0) & (thirteen > 0)).astype(int)
    return df.merge(sdg_agg, on='entity_id', how='left').fillna(0)

feature_df = train.copy()
feature_df['log_revenue'] = np.log1p(feature_df['revenue'].clip(lower=0))
for col in ['overall_score','environmental_score','social_score','governance_score']:
    feature_df[f'{col}_squared'] = feature_df[col] ** 2
feature_df['env_overall_gap'] = feature_df['environmental_score'] - feature_df['overall_score']
feature_df['gov_env_gap'] = feature_df['governance_score'] - feature_df['environmental_score']
feature_df['social_env_gap'] = feature_df['social_score'] - feature_df['environmental_score']
feature_df['env_score_region_resid'] = feature_df['environmental_score'] - feature_df.groupby('region_code')['environmental_score'].transform('mean')
feature_df['log_rev_country_resid'] = feature_df['log_revenue'] - feature_df.groupby('country_code')['log_revenue'].transform('mean')
feature_df = pd.get_dummies(feature_df, columns=['region_code','country_code'], prefix=['region','country'], drop_first=False)
feature_df = add_sector_features(feature_df, sect)
feature_df = add_environmental_features(feature_df, env)
feature_df = add_sdg_features(feature_df, sdg)

feature_df['env_positive_ratio'] = feature_df['env_positive_cnt'] / (feature_df['env_event_cnt'] + 1)
feature_df['env_adj_per_event'] = feature_df['env_adj_sum'] / (feature_df['env_event_cnt'] + 1)
feature_df['has_env_activity'] = (feature_df['env_event_cnt'] > 0).astype(int)
feature_df['has_sdg'] = (feature_df['sdg_count'] > 0).astype(int)
climate_cols = [col for col in ['sdg_7_flag','sdg_12_flag','sdg_13_flag'] if col in feature_df.columns]
feature_df['sdg_climate_focus'] = feature_df[climate_cols].sum(axis=1) if climate_cols else 0
feature_df['climate_focus_x_intensity'] = feature_df['sdg_climate_focus'] * feature_df['sect_high_intensity_pct']
feature_df['sdg_per_logrev'] = feature_df['sdg_count'] / (feature_df['log_revenue'] + 1e-3)
feature_df['logrev_x_highintensity'] = feature_df['log_revenue'] * feature_df['sect_high_intensity_pct']
feature_df['logrev_x_envgap'] = feature_df['log_revenue'] * feature_df['env_overall_gap']
feature_df['intensity_x_envgap'] = feature_df['sect_high_intensity_pct'] * feature_df['env_overall_gap']
feature_df['logrev_x_envposratio'] = feature_df['log_revenue'] * feature_df['env_positive_ratio']
feature_df['sdg_focus_x_envgap'] = feature_df['sdg_climate_focus'] * feature_df['env_overall_gap']
feature_df = feature_df.replace([np.inf, -np.inf], 0).fillna(0)
feature_df.head()
```

```python
drop_cols = ['entity_id','region_name','country_name','target_scope_1','target_scope_2']
numeric_feats = feature_df.drop(columns=drop_cols, errors='ignore').select_dtypes(include=[int, float])
corr_table = pd.DataFrame({
    'feature': numeric_feats.columns,
    'corr_scope1': numeric_feats.corrwith(feature_df['target_scope_1']),
    'corr_scope2': numeric_feats.corrwith(feature_df['target_scope_2']),
})
corr_table['abs_mean_corr'] = corr_table[['corr_scope1','corr_scope2']].abs().mean(axis=1)
corr_table = corr_table.sort_values('abs_mean_corr', ascending=False)
corr_table.head(15)
```

```python
top_feats = corr_table.head(10)
plot_data = top_feats.melt(id_vars='feature', value_vars=['corr_scope1','corr_scope2'], var_name='target', value_name='corr')
fig, ax = plt.subplots(figsize=(10,6))
for idx, target in enumerate(plot_data['target'].unique()):
    subset = plot_data[plot_data['target'] == target]
    ax.barh(subset['feature'] + ' (' + target.split('_')[1] + ')', subset['corr'], alpha=0.75)
ax.axvline(0, color='black', linewidth=1)
ax.set_title('Top Correlated Features with Scope 1 & 2')
ax.set_xlabel('Correlation coefficient')
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12,5), sharey=True)
axes[0].scatter(feature_df['log_revenue'], feature_df['target_scope_1'], alpha=0.6, s=20)
axes[0].set_title('Log-Revenue vs Scope 1')
axes[0].set_xlabel('log(1 + revenue)')
axes[0].set_ylabel('Scope 1 emissions')
axes[1].scatter(feature_df['log_revenue'], feature_df['target_scope_2'], alpha=0.6, s=20, color='tab:orange')
axes[1].set_title('Log-Revenue vs Scope 2')
axes[1].set_xlabel('log(1 + revenue)')
plt.tight_layout()
plt.show()
```

```python
feature_df['high_intensity_flag'] = (feature_df['sect_high_intensity_pct'] > 0.5).astype(int)
fig, ax = plt.subplots(1, 2, figsize=(12,5), sharey=True)
feature_df.boxplot(column='target_scope_1', by='high_intensity_flag', ax=ax[0])
ax[0].set_title('Scope 1 by High-Intensity Mix')
ax[0].set_xlabel('High-intensity share > 50%?')
ax[0].set_ylabel('Scope 1 emissions')
feature_df.boxplot(column='target_scope_2', by='high_intensity_flag', ax=ax[1])
ax[1].set_title('Scope 2 by High-Intensity Mix')
ax[1].set_xlabel('High-intensity share > 50%?')
plt.suptitle('')
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.scatter(feature_df['env_positive_cnt'], feature_df['target_scope_2'], alpha=0.6, color='tab:green', s=20)
ax.set_title('Scope 2 vs Environmentally Positive Actions')
ax.set_xlabel('Count of positive env adjustments (activities)')
ax.set_ylabel('Scope 2 emissions')
plt.tight_layout()
plt.show()
```

## Evidence Recap
- **Scale dominates**: scatter plots show monotonic relationships between log-revenue and both targets, justifying revenue-based transforms and residual features.
- **Sector structure matters**: box plots confirm heavy-industry mixes dramatically shift emission distributions; high-intensity share and concentration metrics are essential.
- **ESG behavior is predictive**: correlation table plus environmental activity scatter reveal strong signals from SDG flags (2, 12) and positive adjustment counts, supporting commitment/activity features.
Together, these visuals validate the engineered feature blocks as the strongest levers for outperforming baseline models.

```python
explanations = {
    'log_revenue': 'Large firms emit more, so a log of revenue captures “size” without letting extreme companies dominate.',
    'log_rev_country_resid': 'Shows whether a company is unusually big relative to peers in the same country—useful because policies and grids vary locally.',
    'logrev_x_highintensity': 'Pairs company scale with how exposed it is to heavy industries; big heavy-mix firms drive emissions the most.',
    'sect_high_intensity_pct': 'Measures what share of revenue comes from carbon-heavy sectors, so it explains structural emitters better than raw sector codes.',
    'sect_C_pct': 'Highlights manufacturing share specifically, a top industrial driver of emissions.',
    'gov_env_gap': 'Tracks how governance outperforms (or lags) environmental scores; governance misalignment often foreshadows under-investment in decarbonization.',
    'governance_score': 'Baseline governance quality still matters; stronger oversight typically yields lower emissions.',
    'env_positive_cnt': 'Counts proactive environmental actions—more projects usually correlate with lower Scope 2 indicators.',
    'env_positive_ratio': 'Normalizes actions by total events to show whether a firm’s actions skew helpful vs. harmful.',
    'logrev_x_envposratio': 'Combines action quality with scale; large firms taking mostly positive actions cut emissions meaningfully.'
}

priority_features = list(explanations.keys())
summary_table = (
    corr_table[corr_table['feature'].isin(priority_features)]
    .copy()
    .sort_values('abs_mean_corr', ascending=False)
    .rename(columns={
        'feature': 'Feature',
        'corr_scope1': 'Correlation vs Scope 1',
        'corr_scope2': 'Correlation vs Scope 2',
        'abs_mean_corr': 'Avg |Correlation|'
    })
)
summary_table['Plain-language reason'] = summary_table['Feature'].map(explanations)
summary_table['Plain-language reason'] = summary_table['Plain-language reason'].apply(lambda text: fill(text, width=55))
summary_table.reset_index(drop=True, inplace=True)
summary_table
```

```python
csv_path = Path('top_feature_story.csv')
summary_table.to_csv(csv_path, index=False)
csv_path
```
