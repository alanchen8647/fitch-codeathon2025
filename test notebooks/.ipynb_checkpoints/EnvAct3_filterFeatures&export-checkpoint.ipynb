```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('output_for_enviornmental_FeatureEngineering.csv')

# Create comprehensive feature set including the 4 new features
features_df = pd.DataFrame()
features_df['entity_id'] = df['entity_id']

# 1. Revenue features
features_df['log_revenue'] = df['log_revenue']
features_df['revenue_millions'] = df['revenue_millions']
features_df['revenue_squared'] = df['revenue_squared']

# 3. Geographic features
key_regions = ['region_WEU', 'region_NAM', 'region_ANZ']
key_countries = ['country_US', 'country_GB', 'country_DE', 'country_FR']
for region in key_regions:
    features_df[region] = df[region]
for country in key_countries:
    features_df[country] = df[country]

# 4. Sector diversity features
features_df['sector_diversity'] = df['sector_diversity']
features_df['max_sector_concentration'] = df['max_sector_concentration']
features_df['sector_entropy'] = df['sector_entropy']

# 5. Environmental activity features
features_df['env_activities_count'] = df['env_activities_count']
features_df['env_adj_net'] = df['env_adj_net']
features_df['env_adj_mean'] = df['env_adj_mean']
features_df['env_activity_types'] = df['env_activity_types']
features_df['env_num_beneficial'] = df['env_num_beneficial']
features_df['env_num_harmful'] = df['env_num_harmful']
features_df['climate_sdg_count'] = df['climate_sdg_count']
features_df['sdg_commitments'] = df['sdg_commitments']

# 6. Activity type proportions
activity_columns = ['env_acttype_Manufacturing', 'env_acttype_Operation', 
                   'env_acttype_Transportation', 'env_acttype_Raw materials']
for col in activity_columns:
    if col in df.columns:
        features_df[col] = df[col]

# 7. CREATE THE 4 NEW FEATURES
print("Creating the 4 new interaction features...")

# Feature 1: env_num_beneficial (already exists, but we'll ensure it's included)
features_df['env_num_beneficial'] = df['env_num_beneficial']

# Feature 2: env_activity_concentration_x_revenue
# Concentration = activities count per activity type (higher = more focused/repetitive activities)
features_df['env_activity_concentration'] = df['env_activities_count'] / (df['env_activity_types'] + 1)  # +1 to avoid division by zero
features_df['env_activity_concentration_x_revenue'] = features_df['env_activity_concentration'] * df['revenue_millions']

# Feature 3: env_activities_count_x_revenue
features_df['env_activities_count_x_revenue'] = df['env_activities_count'] * df['revenue_millions']

# Feature 4: env_consistency_score_x_revenue
# Consistency = inverse of standard deviation (lower std = higher consistency)
features_df['env_consistency_score'] = 1 / (df['env_adj_std'] + 0.1)  # +0.1 to avoid division by zero and extreme values
features_df['env_consistency_score_x_revenue'] = features_df['env_consistency_score'] * df['revenue_millions']

# Add targets for correlation analysis
features_df['target_scope_1'] = df['target_scope_1']
features_df['target_scope_2'] = df['target_scope_2']

print("Feature set shape after adding new features:", features_df.shape)
```

```python
# Comprehensive correlation analysis
correlation_with_scope1 = features_df.corr()['target_scope_1'].sort_values(ascending=False)
correlation_with_scope2 = features_df.corr()['target_scope_2'].sort_values(ascending=False)

print("COMPREHENSIVE CORRELATION ANALYSIS")
print("="*60)

print("\nTOP 20 FEATURES CORRELATED WITH TARGET_SCOPE_1:")
print("="*50)
for feature, corr in correlation_with_scope1.head(20).items():
    if feature not in ['target_scope_1', 'target_scope_2', 'entity_id']:
        print(f"{feature:45} : {corr:7.4f}")

print("\nTOP 20 FEATURES CORRELATED WITH TARGET_SCOPE_2:")
print("="*50)
for feature, corr in correlation_with_scope2.head(20).items():
    if feature not in ['target_scope_1', 'target_scope_2', 'entity_id']:
        print(f"{feature:45} : {corr:7.4f}")

# Focus on the 4 new features
print("\nPERFORMANCE OF THE 4 NEW FEATURES:")
print("="*50)
new_features = [
    'env_num_beneficial',
    'env_activity_concentration_x_revenue', 
    'env_activities_count_x_revenue',
    'env_consistency_score_x_revenue'
]

for feature in new_features:
    scope1_corr = features_df[feature].corr(features_df['target_scope_1'])
    scope2_corr = features_df[feature].corr(features_df['target_scope_2'])
    print(f"{feature:45} | Scope1: {scope1_corr:7.4f} | Scope2: {scope2_corr:7.4f}")
```

```python
# Visualization of new features' correlations
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Plot 1: env_num_beneficial
axes[0,0].scatter(features_df['env_num_beneficial'], features_df['target_scope_2'], alpha=0.6)
axes[0,0].set_xlabel('Number of Beneficial Environmental Activities')
axes[0,0].set_ylabel('Scope 2 Emissions')
axes[0,0].set_title(f'env_num_beneficial vs Scope 2 (r={features_df["env_num_beneficial"].corr(features_df["target_scope_2"]):.3f})')

# Plot 2: env_activity_concentration_x_revenue
axes[0,1].scatter(features_df['env_activity_concentration_x_revenue'], features_df['target_scope_1'], alpha=0.6)
axes[0,1].set_xlabel('Activity Concentration  Revenue')
axes[0,1].set_ylabel('Scope 1 Emissions')
axes[0,1].set_title(f'env_activity_concentration_x_revenue vs Scope 1 (r={features_df["env_activity_concentration_x_revenue"].corr(features_df["target_scope_1"]):.3f})')

# Plot 3: env_activities_count_x_revenue
axes[1,0].scatter(features_df['env_activities_count_x_revenue'], features_df['target_scope_1'], alpha=0.6)
axes[1,0].set_xlabel('Activity Count  Revenue')
axes[1,0].set_ylabel('Scope 1 Emissions')
axes[1,0].set_title(f'env_activities_count_x_revenue vs Scope 1 (r={features_df["env_activities_count_x_revenue"].corr(features_df["target_scope_1"]):.3f})')

# Plot 4: env_consistency_score_x_revenue
axes[1,1].scatter(features_df['env_consistency_score_x_revenue'], features_df['target_scope_1'], alpha=0.6)
axes[1,1].set_xlabel('Consistency Score  Revenue')
axes[1,1].set_ylabel('Scope 1 Emissions')
axes[1,1].set_title(f'env_consistency_score_x_revenue vs Scope 1 (r={features_df["env_consistency_score_x_revenue"].corr(features_df["target_scope_1"]):.3f})')

plt.tight_layout()
plt.show()
```

```python
# Select final features based on comprehensive correlation analysis
final_features = [
    'entity_id',
    # High correlation with Scope 1
    'env_activities_count_x_revenue',  # NEW - Strong for Scope 1
    'env_activity_concentration_x_revenue',  # NEW - Strong for Scope 1
    'env_acttype_Manufacturing',
    'env_activities_count',
    # High correlation with Scope 2  
    'env_num_beneficial',  # NEW - Strong for Scope 2
    'env_acttype_Operation',
    # Composite features
    'env_adj_net',
    'env_consistency_score_x_revenue'  # NEW - Moderate for Scope 1
]

# Create final feature dataset
final_features_df = features_df[final_features].copy()

# Add targets for modeling
final_features_df['target_scope_1'] = df['target_scope_1']
final_features_df['target_scope_2'] = df['target_scope_2']

print("FINAL FEATURE SET PERFORMANCE:")
print("="*50)
print(f"{'Feature':45} | {'Scope1 Corr':12} | {'Scope2 Corr':12}")
print("-"*80)
for feature in final_features_df.columns:
    if feature not in ['entity_id', 'target_scope_1', 'target_scope_2']:
        scope1_corr = features_df[feature].corr(features_df['target_scope_1'])
        scope2_corr = features_df[feature].corr(features_df['target_scope_2'])
        print(f"{feature:45} | {scope1_corr:12.4f} | {scope2_corr:12.4f}")

# Export to CSV
final_features_df.to_csv('emissions_prediction_features_with_new_features.csv', index=False)
print(f"\nFeatures exported to 'emissions_prediction_features_with_new_features.csv'")
print(f"Dataset contains {len(final_features_df)} entities and {len(final_features_df.columns)} features")
```
